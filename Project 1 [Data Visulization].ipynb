```python
class RandomWalk:
    def __init__(self,num_points =5000):
        self.num_points = num_points

        self.x = [0]
        self.y = [0]

    def fill_walk(self):
        while len(self.x) < self.num_points:
            x_direction = choice([1,-1])
            x_distance = choice([0,1,2,3,4])
            x_step = x_direction * x_distance
            
            y_direction= choice([1,-1])
            y_distance = choice([0,1,2,3,4])
            y_step = y_direction * y_distance
            
            if x_step == 0 and y_step == 0:
                continue
                
#xf is for x final value

            xf = self.x[-1] + x_step
            yf = self.y [-1] + y_step
            
            self.x.append(xf)
            self.y.append(yf)
            

rw = RandomWalk()

rw.fill_walk()

plt.style.use('classic')
plt.scatter(rw.x,rw.y, s= 10)

plt.show()
```

```python
class ChosingRandomPoints:
    def __init__(self, num_of_points = 1000):
        self.num_of_points = num_of_points
        self.x = [0]
        self.y = [0]
        
    def filling_random_points(self):
        while len(self.x) < self.num_of_points:
            
            x_direction = choice([1,-1])
            x_distance = choice([0,2,4,6,8,10])
            x_stepping = x_direction * x_distance
            
            y_direction = choice([1,-1])
            y_distance = choice([0,1,2,3,4,5])
            y_stepping = y_direction * y_distance
            
            if x_stepping and y_stepping == 0 :
                continue
            
            xf = self.x [-1] + x_stepping
            yf = self.y [-1] + y_stepping
            
            self.x.append(xf)
            self.y.append(yf)
            

while True:
    crp = ChosingRandomPoints()
    crp.filling_random_points()


    fig,ax = plt.subplots(figsize=(15,12))

    ax.set_title('Chosing Random Points Within 1000', fontsize = 15)
    ax.set_xlabel('X Axis', fontsize = 12)
    ax.set_ylabel('Y Axis', fontsize = 12)
    ax.scatter(crp.x,crp.y, cmap= plt.cm.Blues, s = 30)

    plt.show()
    
    repeat = input('Want to repeat? : ')
    if repeat == 'no' or 'n' or 'negative':
        break
```

```python
from random import randint
from plotly.graph_objs import Bar, Layout
class Die:
    def __init__(self, num_sides = 6):
        self.num_sides = num_sides
        
    def roll(self):
        return randint(1, self.num_sides)
    
die= Die()
results = []
for roll_num in range(1000):
    result = die.roll()
    results.append(result)
    
frequencies = []
for value in range(1,die.num_sides+1 ):
    frequency = results.count(value)
    frequencies.append(frequency)


xv = list(range(1, die.num_sides + 1))
data = [Bar(x = xv,y = frequencies)]

x_axis_config = {'title': 'Results'}
y_axis_config = {'title': 'Frequency of Result'}
my_layout = Layout(title = 'Results of Rolling one D6 100 Times', xaxis = x_axis_config, yaxis = y_axis_config)

offline.plot({'data' : data, 'layout' : my_layout}, filename = 'd6.html')
```

```python
def r_1_6():
    '''Returning numbers between 1 to 6,it is like taking randomly
    '''
    return randint(1,6)

def l_return():
    '''Looping the returning numbers 1 to 6 for 2000 times
    '''
    our_f_list = []    
    '''our final list'''
    
    for rolling_num in range(2000):
        l_looping = r_1_6()       
        '''list looping'''
        
        our_f_list.append(l_looping)
        

    def fhmtdhr():
        '''finding how much the dice has rolled
        '''
        frqs = []
        for value in range(1,7):
            frq = our_f_list.count(value)
            '''Frequency
            '''
            frqs.append(frq)
            '''Frequencies
            '''
        
        xv = list(range(1,7))
        data = [Bar(x = xv, y = frqs)]
        
        xac = {'title': 'Results of Rolled'}
        yac = {'title': 'Frequency of Results'}
        
        ml = Layout(title = 'Finding the results of rolling one dice 2000 times', xaxis = xac, yaxis = yac)
        
        offline.plot({'data' : data, 'layout' : ml}, filename= 'd6(2000).html')
        
    fhmtdhr()       
l_return()
```

```python
from plotly.graph_objs import Scattergeo, Layout

filename = 'eq_data_30_day_m1.json'
with open(filename) as f:
    one_month_earthqukes = json.load(f)
    
new_filename = 'readable_30_days_data.json'
with open(new_filename,'w') as nf:
    json.dump(one_month_earthqukes, nf, indent=4)
    
    
all_data = one_month_earthqukes ['features']

mags, longitudes, latitudes = [], [] , []

for dictionary in all_data:
    mag = dictionary['properties']['mag']
    longitude = dictionary ['geometry']['coordinates'][0]
    latitude = dictionary ['geometry'] ['coordinates'][1]
    
    mags.append(mag)
    longitudes.append(longitude)
    latitudes.append(latitude)
    
# Mapping the earthquakes

data = [{
    'type' : 'scattergeo',
    'lon' : longitudes,
    'lat' : latitudes,
    'marker' : {
        'size' : [5*mag for mag in mags],
        'color' : mags,
        'colorscale' : 'Virdis',
        'reversescale' : True,
        'colorbar' : { 'title' : 'Magnitude'},
    },
}]

ml = Layout(title = 'Global Earthquakes in Last Month') 


fig = {'data': data, 'layout': ml}

offline.plot(fig, filename = 'Last Month Earthquakes in the World.html')
```
